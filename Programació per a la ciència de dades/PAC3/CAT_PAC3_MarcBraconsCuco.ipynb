## Exercici 1
El número $\pi$ es pot aproximar amb la següent sèrie de sumands:

    pi = 4 (1-1/3+1/5-1/5+1/7-1/7+1/9...)

només cal donar més valors a la sèrie, i obtindrem una millor aproximació. Aquesta sèrie la podem implementar amb la següent funció:

    def approximate_pi(N):
        value = 0
        for k in range(1, 2*N+1, 2):
            sign = -(k % 4 - 2)
            value += float(sign) / k
        return 4 * value

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **(1 punt)**

### 1.1
Compta els passos que realitza la funció **approximate_pi()** i dedueix-ne la seva complexitat assimptòtica en el pitjor cas (notació $\mathcal{O}$), a partir del número de passos que s'executen.

**Solució 1.1**.

- Inicialitzem la variable 'value': Això només es fa un cop independentment de N.
- Bucler 'for': iterem de 1 fins a 2*N+1 amb increments de 2 això vol dir que farem N iteracions.
    - 'sign = -(k % 4 - 2)' es fa a temps constant
    - 'value += float(sign) / k' es fa a temps constant
- return '4 * value' es fa a temps constant

El factor que requereix més temps és el bucle 'for'. Com aquest s'executa N vegades, la complexitat assimptòtica en el pitjor cas és $\mathcal{O}(N)$ (algorisme lineal).

### 1.2
Confirma aquest resultat fent una gràfica del temps que triga a executar-se l'algorisme amb diferents valors de número de sumands. Pots provar fins a 10000000.

**Solució 1.2**

Confirmem el resultat esperat, en la gràfica podem veure el comportament lineal

```python
import matplotlib.pyplot as plt

# Definició de l'aproximació
def approximate_pi(N):
    value = 0
    for k in range(1, 2 * N + 1, 2):
        sign = -(k % 4 - 2)
        value += float(sign) / k
    return 4 * value

# Valors de N a provar
N_values = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
execution_times = []

# Mesurem el temps d'execució
for N in N_values:
    t = timeit.Timer(lambda: approximate_pi(N))
    # Cridem la funció 10 vegades en cada prova, repetim 5 vegades
    min_time = min(t.repeat(number=10, repeat=5))
    execution_times.append(min_time)

plt.plot(N_values, execution_times, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Nombre de Sumands (N)')
plt.ylabel('Temps d\'Execució (segons) [Mínim]')
plt.title('Temps d\'Execució de approximate_pi(N)')
plt.grid(True)
plt.show()
```

### 2.2
Volem obtenir dígits semi-aleatoris de 6 xifres. El que fareu és llegir en el fitxer pi.txt blocs de 6 bytes. Aquests blocs els hauràs primer de decodificar per convertir-los a string, i després convertir-los a números (int). Posa tots aquests números en una llista. Com que hi haurà números repetits, elimina els repetits. Quants números hi ha a la llista? Quin és el primer? Quin és el més petit? Quin és el més gran?

**Solució 2.2**

Definim dues funcions una per llegir el fitxer en blocs de 6 digits i l'altre per eliminar duplicats (aquesta la fem per a mantindre l'ordre de la llista, si l'ordre no importa podriem fer 'list(set(hola))', ja que 'set' no permet elements repetits però al fer-ho perdem l'ordre.
- funció 'llegir_blocs_6_bytes' és una modificació de la funció de l'apartat 2.1
    - En comptes de moure el cursor a la posició inicial, el fem servir per avançar al següent bloc
    - Ho fa fins que llegeix un bloc buit (ja estem al final del fitxer)

- funció 'eliminar_duplicats_mantenint_ordre': la fem servir per eliminar duplicats d'una llista mantenint l'ordre
    - La idea és semblant a fer directament 'list(set(llista)) però ho fem en un conjunt auxiliar per a mantindre l'ordre.

Fem servir 4 threads per a dividir la tasca de lectura. Al final del codi s'imprimeixen els resultats que demana l'exercici.

```python
from typing import List


def llegir_blocs_6_bytes(ruta_fitxer: str, inici: int, dimensio: int,
                         lock: threading.Lock, numeros: List[int]) -> None:
    """
    Llegeix blocs de 6 bytes des d'un fitxer, converteix cada bloc en un enter i
    l'afegeix a una llista

    Args:
    ruta_fitxer (str): Ruta al fitxer binari a llegir.
    inici (int): Posició inicial dins el fitxer des d'on començar a llegir.
    dimensio (int): Nombre de bytes a llegir des de la posició inicial.
    lock (Lock): Objecte de bloqueig per controlar l'accés a la llista compartida.
    numeros (List[int]): Llista on s'afegeixen els nombres llegits.
    """
    with open(ruta_fitxer, 'rb') as fitxer:
        # movem cursor a 'inici'
        fitxer.seek(inici)
        # calculem la posició final
        final = inici + dimensio
        while fitxer.tell() < final:
            bloc = fitxer.read(6)
            if not bloc:
                break
            try:
                # convertim cada bloc en un enter
                numero = int(bloc.decode('utf-8'))
                # ho fem per a que no hi hagi conflictes 
                # en l'accés concurrent a la llista
                with lock:
                    # afegim a la llista
                    numeros.append(numero)
            # si hi ha algun error el captura i el bucle continua
            except Exception:
                continue


def eliminar_duplicats_mantenint_ordre(numeros: List[int]) -> List[int]:
    """
    Elimina els duplicats d'una llista mantenint l'ordre original dels elements.

    Args:
    numeros (List[int]): Llista de nombres des d'on eliminar duplicats.

    Returns:
    List[int]: Nova llista sense duplicats.
    """
    # llista 'numeros_unics' per anar guardant els números únics
    # set 'vistos' per mirar elements ja afegits a numeros_unics
    numeros_unics = []
    vistos = set()
    # per a cada número mirem si està a 'vistos'.
    # Si no hi és, l'afegim a 'vistos' i 'numeros_unics'
    # manetenim l'ordre perquè estem recorrent la llista en ordre
    for numero in numeros:
        if numero not in vistos:
            numeros_unics.append(numero)
            vistos.add(numero)
    return numeros_unics

# nombre de threads
num_threads = 4
# la mida de cada thread és el quocient entre
# la mida del fitxer i el nombre de threads
mida_per_thread = mida_fitxer // num_threads

# llista per guardar números
numeros = []
# el fem servir per evitar problemes quan es modifica 'numeros'
lock = threading.Lock()
# llista per guardar objectes de thread
threads = []

# Creem i iniciem els fils
for i in range(num_threads):
    inici = i * mida_per_thread
    thread = threading.Thread(target=llegir_blocs_6_bytes,
                              args=(ruta_fitxer, inici, mida_per_thread,
                                    lock, numeros))
    threads.append(thread)
    thread.start()

# Esperem a que acabin
for thread in threads:
    thread.join()

# Eliminem duplicats
numeros_unics = eliminar_duplicats_mantenint_ordre(numeros)

# Resultats
print(f"Quantitat de números únics: {len(numeros_unics)}")
if numeros_unics:
    print(f"Primer número: {numeros_unics[0]}")
    print(f"Número més petit: {min(numeros_unics)}")
    print(f"Número més gran: {max(numeros_unics)}")
```

```python
# mètode Sentinel Linear Search
def sentinelSearch(arr, n, key):
    """
    Python3 implementation of the approach
    Function to search key in the given array
    This code is contributed by divyamohan123, Mandeep Dalavi
    """ 
    # Last element of the array
    last = arr[n - 1]
 
    # Element to be searched is
    # placed at the last index
    arr[n - 1] = key
    i = 0
 
    while (arr[i] != key):
        i += 1
 
    # Put the last element back
    arr[n - 1] = last
 
    if ((i < n - 1) or (arr[n - 1] == key)):
        return i
    else:
        return -1

        
result = sentinelSearch(numeros_unics, N, numeros_a_buscar[0])
# assert del número que si que hi hauria de ser
assert result != -1, f'El número {numeros_a_buscar[0]} no està a la llista'
print(f'El número {numeros_a_buscar[0]} està a la llista i el seu índex és {result}')

result = sentinelSearch(numeros_unics, N, numeros_a_buscar[1])
# assert del número que no hi hauria de ser
assert result == -1, f'El número {numero} està a la llista'
print(f'El número {numeros_a_buscar[1]} no està a la llista')
```

## 3.2

**Solució 3.2**

**Solució 4.1**

Si dividim la posició dels element de la llista en parts iguals de forma que totes sumades tenen valor 100, estem buscant l'element tal que la seva posició sumada a tots els elements més petit que ell, suma 73.

**Comparació en esacala normal:**
El mètode Search augmenta amb la longitud de la llista ja que té una complexitat $\mathcal{O}(N)$, on N és la longitud de la llista.

En canvi, el mètode Binary té un temps d'execusió constant i independent de la longitud. Això és perquè té una complexitat $\mathcal{O}(log(N))$ (necessita un nombre logaritmic de passos per trobar la key o dir que no hi és).

**Comparació en escala logarítmica:**
L'escala logarítmica ens permet veure millor les diferències en ordres de magnitud.

El mètode Binary és més ràpid amb un temps d'execussió pràcticament horitzontal

**En resum:**
El mètode Binary és més ràpid però necessitem que la llista estigui ordenada.

```python
def binarySearch(arr, l, r, x):
    """
    Python3 code to implement iterative Binary Search.
    Source: https://www.geeksforgeeks.org/binary-search/
    """
    while l <= r:
        mid = l + (r - l) // 2
        # Check if x is present at mid
        if arr[mid] == x:
            return mid
        # If x is greater, ignore left half
        elif arr[mid] < x:
            l = mid + 1
        # If x is smaller, ignore right half
        else:
            r = mid - 1
    # If we reach here, then the element was not present
    return -1

# Igual que abans però amb 'numeros_ordenats' i comença a partir del 100
longituds_llista = np.linspace(100, len(numeros_ordenats), 100, dtype=int)

temps_lineal = []
temps_binary = []

for longitud in longituds_llista:
    arr = numeros_unics[:longitud]
    index = int(len(arr) * 0.73)
    key = arr[index]

    t_lineal = timeit.Timer(lambda: search(arr, longitud, key))
    temps_lineal.append(min(t_lineal.repeat(number=6, repeat=3)))

    t_binary = timeit.Timer(lambda: binarySearch(arr, 0, longitud - 1, key))
    temps_binary.append(min(t_binary.repeat(number=6, repeat=3)))

# Gràfic en escala normal
plt.figure(figsize=(12, 6))
plt.plot(longituds_llista, temps_lineal, label='Lineal Bàsica')
plt.plot(longituds_llista, temps_binary, label='Binària', linestyle='--')
plt.xlabel('Longitud de la Llista')
plt.ylabel('Temps d\'Execució (segons)')
plt.title('Comparació de Temps d\'Execució: Lineal vs Binària')
plt.legend()
plt.grid(True)
plt.show()

# Gràfic en escala logarítmica
plt.figure(figsize=(12, 6))
plt.plot(longituds_llista, temps_lineal, label='Lineal Bàsica')
plt.plot(longituds_llista, temps_binary, label='Binària', linestyle='--')
plt.xlabel('Longitud de la Llista')
plt.ylabel('Temps d\'Execució (segons)')
plt.title('Comparació de Temps d\'Execució en Escala Logarítmica: Lineal vs Binària')
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

### 4.2
Implementa els tres primers algorismes de la llista: Binary Search, Meta Binary Search i Ternary Search. Posa en una mateixa gràfica per tal d'observar si hi ha alguna diferència significativa. Recorda que aquests algorismes són de tipus log(N), i per tant si poses escala logarítmica en el temps d'execució hauries de veure un comportament lineal.

**Solució 4.2**

Sobre la gràfica:
- Podem veure que el mètode 'Ternary' és millor per a llistes menys de 1000 valors. Entre els 100 i els 1000 valors hi ha un punt en el que els altres mètodes passen a tindre una eficacia millor.
- El mètode 'bsearch' es comporta igual que el 'Binary' però sempre en temps més baixos. 
- Una cosa interessant és que es veu un comportament lineal en els tres mètodes fins a arribar a llistes de més de 3000 elements, a partir d'aquí hi ha molts pics.

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
import math as mt

def bsearch(A, key_to_search):
    """
    This implementation was improved by Tanin
    This code is contributed by ChitraNayal
    """
    n = len(A)
    lg = int(mt.log2(n - 1)) + 1

    pos = 0
    for i in range(lg - 1, -1, -1):
        if A[pos] == key_to_search:
            return pos

        new_pos = pos | (1 << i)
        if (new_pos < n) and (A[new_pos] <= key_to_search):
            pos = new_pos

    return pos if A[pos] == key_to_search else -1


def ternary_search(l, r, key, ar):
    """
    This code is contributed by Mohit kumar 29
    """
    if r >= l:
        mid1 = l + (r - l) // 3
        mid2 = r - (r - l) // 3

        if ar[mid1] == key:
            return mid1
        if ar[mid2] == key:
            return mid2

        if key < ar[mid1]:
            return ternary_search(l, mid1 - 1, key, ar)
        elif key > ar[mid2]:
            return ternary_search(mid2 + 1, r, key, ar)
        else:
            return ternary_search(mid1 + 1, mid2 - 1, key, ar)

    return -1

longituds_llista = np.linspace(100, len(numeros_ordenats), 100, dtype=int)

# Aquí guardarem els temps
temps_bsearch = []
temps_binary = []
temps_ternary = []

# Mesurem dels temps d'execució
for longitud in longituds_llista:
    arr = numeros_ordenats[:longitud]
    index = int(len(arr) * 0.73)
    key = arr[index]

    # bsearch
    t_bsearch = timeit.Timer(lambda: bsearch(arr, key))
    temps_bsearch.append(min(t_bsearch.repeat(number=10, repeat=5)))

    # binarySearch
    t_binary = timeit.Timer(lambda: binarySearch(arr, 0, longitud - 1, key))
    temps_binary.append(min(t_binary.repeat(number=10, repeat=5)))

    # ternarySearch
    t_ternary = timeit.Timer(lambda: ternary_search(0, longitud - 1, key, arr))
    temps_ternary.append(min(t_ternary.repeat(number=10, repeat=5)))

# Gràfic amb escala logarítmica
plt.figure(figsize=(12, 6))
plt.plot(longituds_llista, temps_bsearch, label='bsearch')
plt.plot(longituds_llista, temps_binary, label='Binary')
plt.plot(longituds_llista, temps_ternary, label='Ternary', linestyle='-.')
plt.xlabel('Longitud de la Llista')
plt.ylabel('Temps d\'Execució (segons)')
plt.title('Comparació de Temps d\'Execució en Escala Logarítmica')
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

**Solució 5.1**

```python
import decimal
import math


def compute_pi(n):
    """
    This function calculates the value of pi to 'n' number of decimal places
    Args:
    n:   precision(Decimal places)
    Returns:
    pi:   the value of pi to n-decimal places
    """

    decimal.getcontext().prec = n + 3
    decimal.getcontext().Emax = 999999999

    C = 426880 * decimal.Decimal(10005).sqrt()
    K = decimal.Decimal(6)
    M = decimal.Decimal(1)
    X = decimal.Decimal(1)
    L = decimal.Decimal(13591409)
    S = L

    # For better precision, we calculate to n+3 and truncate the last two digits
    for i in range(1, n + 3):
        M = decimal.Decimal(
            M * ((1728 * i * i * i) - (2592 * i * i) + (1104 * i) - 120) / (i * i * i)
        )
        L = decimal.Decimal(545140134 + L)
        X = decimal.Decimal(-262537412640768000 * X)
        S += decimal.Decimal((M * L) / X)

    return str(C / S)[:-2]  # Pi is C/S


n = 100
print(f"Primers {n} decimals de Pi \n")
print(compute_pi(n))
```

**Solució 6.2**

Hem modificat la funció de l'apartat 6.1 per a que només en quedi la part de descarregar fitxers. També hem fet una funció que executa 'tasca_descarregar' i 'compute_pi' en threads diferents.

```python
def tasca_descarregar():
    pdf_urls = [
        "https://www.cs.umb.edu/~offner/files/pi.pdf",
        "https://www.davidhbailey.com/dhbpapers/pi-formulas.pdf",
        "https://www.davidhbailey.com/dhbtalks/dhb-conant.pdf",
        "https://arxiv.org/pdf/1912.05319.pdf",
        "https://maa.org/sites/default/files/pdf/pubs/amm_supplements/Monthly_Reference_12.pdf",
    ]

    for i, url in enumerate(pdf_urls, start=1):
        descarregar_fitxer(url, f"document_{i}.pdf")


def tasques_en_parallel():
    calcul_thread = threading.Thread(target=compute_pi, args=(7000,))
    descarregar_thread = threading.Thread(target=tasca_descarregar)

    calcul_thread.start()
    descarregar_thread.start()

    calcul_thread.join()
    descarregar_thread.join()


t = timeit.Timer(tasques_en_parallel)

times = t.repeat(number=2, repeat=2)
min_time_6_2 = min(times)
print(
    f"El temps mínim per calcular 7000 decimals de pi i descarregar els PDFs és: {min_time_6_2} segons"
)
```

### 7.1
En un procés calcularem els 7000 decimals del número pi. I també crearem un procés per cada descàrrega de fitxer. Igual que abans, volem comparar la versió amb un sol procés (ja calculat) amb la versió multiprocés. Calcula el temps total.

**Solució 7.1**

```python
import multiprocessing
import timeit

pdf_urls = [
    "https://www.cs.umb.edu/~offner/files/pi.pdf",
    "https://www.davidhbailey.com/dhbpapers/pi-formulas.pdf",
    "https://www.davidhbailey.com/dhbtalks/dhb-conant.pdf",
    "https://arxiv.org/pdf/1912.05319.pdf",
    "https://maa.org/sites/default/files/pdf/pubs/amm_supplements/Monthly_Reference_12.pdf",
]


def tasca_descarregar(url, filename):
    descarregar_fitxer(url, filename)


def tasques_en_parallel():
    process_calcul = multiprocessing.Process(target=compute_pi, args=(7000,))
    download_processes = []

    for i, url in enumerate(pdf_urls):
        process = multiprocessing.Process(
            target=tasca_descarregar, args=(url, f"document_{i}.pdf")
        )
        download_processes.append(process)

    process_calcul.start()
    for process in download_processes:
        process.start()

    process_calcul.join()
    for process in download_processes:
        process.join()


t = timeit.Timer(tasques_en_parallel)
times = t.repeat(number=2, repeat=2)
min_time_7_1 = min(times)
print(
    f"El temps mínim per calcular 7000 decimals de pi i descarregar els PDFs és: {min_time_7_1} segons"
)
```
